```python
import numpy as np

def calculate_rmsd(coords1, coords2):
    diff = coords1 - coords2
    rmsd = np.sqrt(np.mean(np.sum(diff ** 2, axis=1)))
    return rmsd

# Function to read xyz file with repeating pattern
def read_xyz_file(file_path):
    with open(file_path, 'r') as f:
        lines = f.readlines()
        num_atoms = int(lines[0])
        coords_list = []
        start_index = 2
        while start_index < len(lines):
            coords = []
            for line in lines[start_index:start_index + num_atoms]:
                coords.append([float(x) for x in line.split()[1:]])
            coords_list.append(np.array(coords))
            start_index += num_atoms + 2
    return coords_list


# Read xyz file
xyz_file = 'alanines_aligned.xyz'
coords_list = read_xyz_file(xyz_file)

# Calculate average coordinates
average_coords = np.mean(coords_list, axis=0)

# Calculate RMSD for each structure from the average
rmsd_values = []
for coords in coords_list:
    rmsd = calculate_rmsd(coords, average_coords)
    rmsd_values.append(rmsd)
```

```python
# Create a DataFrame for Seaborn
data = pd.DataFrame({'RMSD': rmsd_values})
data = data[data['RMSD'] < 50]

plt.figure(figsize=(8, 5), dpi=300)

sns.histplot(data['RMSD'], kde=True, bins=25, color=purple, alpha=0.65)

mean=data['RMSD'].mean()
print(mean)
plt.axvline(data['RMSD'].mean(), color=purple, linestyle='dashed', linewidth=1, label=f"Mean: {mean:.2f}")

plt.xlabel('RMSD (Ã…)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks([1.8, 2.2, 2.6, 3.0], minor=True,fontsize=12,)
plt.yticks(fontsize=12)


plt.tight_layout()
plt.savefig('rmsd_distribution.png', bbox_inches='tight', dpi=300)

plt.show()
```
